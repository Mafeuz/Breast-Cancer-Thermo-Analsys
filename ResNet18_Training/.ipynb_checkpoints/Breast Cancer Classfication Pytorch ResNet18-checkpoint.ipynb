### Training Setup

```python
# Config Dataset and Preprocessing:
data_path  = 'C:\\Users\\MATHEUS DANTAS PEREI\\Desktop\\mastology_dataset_ready\\masto_dataset\\'
img_resize = 100
batch_size = 8
workers    = 2
device     = 'cuda:0'

# Train Config:
learning_rate  = 0.00045
momentum       = 0.9
lr_decay_gamma = 0.1
lr_decay_step  = 7
num_epochs     = 350
save_path      = 'C:\\Users\\MATHEUS DANTAS PEREI\\Desktop\\mastology_dataset_ready\\ResNet18\\mastology_models\\'

# Dataloaders:
DATA = resnet18_utils.resnet18_data_preprocessing(data_path, img_resize, batch_size, workers, device)
dataloaders, dataset_sizes, class_names, device = DATA

# ResNet18  Fit Config:
model_ft    = models.resnet18(pretrained=True)
num_ftrs    = model_ft.fc.in_features
model_ft.fc = nn.Linear(num_ftrs, len(class_names))
model_ft    = model_ft.to(device)
criterion   = nn.CrossEntropyLoss()

# Parameters optimization:
optimizer_ft = optim.SGD(model_ft.parameters(), lr=learning_rate, momentum=momentum)

# Learning Rate Decay by a factor of gamma every step_size epochs:
lr_scheduler = lr_scheduler.StepLR(optimizer_ft, step_size=lr_decay_step, gamma=lr_decay_gamma)

# Training, model fit:
model_ft = resnet18_utils.train_model(model_ft, 
                                      dataloaders,
                                      dataset_sizes,
                                      device, 
                                      save_path, 
                                      criterion, 
                                      optimizer_ft, 
                                      lr_scheduler, 
                                      num_epochs)
```

### Inference and Testing:

```python
from PIL import Image
import glob

# Data path:
data_paths = ['C:\\Users\\MATHEUS DANTAS PEREI\\Desktop\\mastology_dataset_ready\\masto_dataset\\test\\healthy\\',
              'C:\\Users\\MATHEUS DANTAS PEREI\\Desktop\\mastology_dataset_ready\\masto_dataset\\test\\sick\\']

# Loading Model:
model_path = 'C:\\Users\\MATHEUS DANTAS PEREI\\Desktop\\mastology_dataset_ready\\ResNet18\\mastology_models\\best_ep8.pt'
classes    = ['healthy', 'sick']
device     = 'cuda:0' 
input_size = 100
batch_size = 8

model, pre_process = resnet18_utils.load_ResNet18(model_path, classes, input_size, device)

# Count Detections:
correct_preds = 0
missed_preds  = 0
n_imgs        = 0
true_labels   = []
pred_labels   = []

# Loop over data paths:
for data_path in data_paths:

    # Get imgs list within path:
    imgs_paths = glob.glob(data_path + '*.jpg')
    
    # Loop over imgs:
    for img_path in imgs_paths:
        
        n_imgs += 1
        class_det, outputs = resnet18_utils.ResNet18_inference(img_path, 
                                                               model, 
                                                               classes, 
                                                               pre_process, 
                                                               batch_size, 
                                                               device)

        ##################################################################################
        # Count detections:
        if (class_det == classes[0]) and (classes[0] in img_path):
            correct_preds += 1
            real_class = classes[0]
            
        if (class_det == classes[1]) and (classes[0] not in img_path):
            correct_preds += 1
            real_class = classes[1]
            
        if (class_det == classes[1]) and (classes[0] in img_path):
            missed_preds += 1
            real_class = classes[0]
            
        if (class_det == classes[0]) and (classes[0] not in img_path):
            missed_preds += 1
            real_class = classes[1]
            
        true_labels.append(real_class)
        pred_labels.append(class_det)
        
        print('Number of Correct Predictions:', correct_preds)
        print('Number of Incorrect Predictions:', missed_preds) 
        print(f'Actual Class: {real_class}')
        ##################################################################################
                
        # Show current img:
        img = cv2.imread(img_path)
        fig = plt.figure(figsize=(4,4))
        plt.imshow(img)
        plt.pause(2)
        #clear_output(wait=False)
        ##################################################################################
        
# Print out results:
print('Number of Correct Predictions:', correct_preds)
print('Number of Incorrect Predictions:', missed_preds)
print('Total of imgs', n_imgs)
```

### Model Evaluation:

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Apply confusion matrix to true_labels and pred_labels colected:
cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize = (10,8))
ax = plt.subplot()
sns.heatmap(cm, annot=True);
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['healthy', 'sick']);
ax.yaxis.set_ticklabels(['healthy', 'sick']);
```

```python
# Get metrics from previous test:

from sklearn.metrics import classification_report
print(classification_report(true_labels, pred_labels, target_names=['healthy', 'sick']))
```
